```python
import pandas as pd
import json

def read_metrics_file(filename):
    data = []
    with open(filename, 'r') as file:
        while True:
            network_metrics = file.readline().strip()
            if not network_metrics: break  # End of file
            results = json.loads(file.readline().strip())
            
            network_path = network_metrics.split('Network: ')[1].split(', Metrics: ')[0]
            metrics = network_metrics.split('Metrics: ')[1]
            
            record = {
                'network': network_path,
                'metrics': metrics,
                **results
            }
            data.append(record)
    
    return pd.DataFrame(data)

# Usage
df = read_metrics_file(results_path)
print(df)
```
