```python
# Set paths
real_images_path = f"{DATA_PATH}/Zalando_Germany_Dataset/dresses/images/e4e_images/all/"
reconstruction_path_00003 = f"{DATA_PATH}/Generated_Images/e4e/00003_snapshot_920/"
reconstruction_path_00005 = f"{DATA_PATH}/Generated_Images/e4e/00005_snapshot_1200/"

real_images = glob(real_images_path+'*.jpg')
reconstructions_00003 = glob(f"{reconstruction_path_00003}*.jpg")
reconstructions_00005 = glob(f"{reconstruction_path_00005}*.jpg")
print(len(real_images), len(reconstructions_00003), len(reconstructions_00005))
```

```python
NUM_SAMPLES = 20
samples = np.random.choice(real_images, NUM_SAMPLES)

fig, ax = plt.subplots(NUM_SAMPLES, 3, figsize = (15,NUM_SAMPLES*5))
for i in range(NUM_SAMPLES):
    path = samples[i]
    sku = path.split('/')[-1].split('.')[0]
    path_00003 = [elem for elem in reconstructions_00003 if sku in elem][0]
    path_00005 = [elem for elem in reconstructions_00005 if sku in elem][0]

    ax[i, 0].imshow(Image.open(path))
    ax[i, 1].imshow(Image.open(path_00003))
    ax[i, 2].imshow(Image.open(path_00005))

    if i%1 == 0:
        ax[i, 0].set_title('Original')
        ax[i, 1].set_title('e4e 00003 Snapshot 920')
        ax[i, 2].set_title('e4e 00005 Snapshot 1200')

for axs in ax.flatten():
    axs.axis('off') 

plt.show()
```
